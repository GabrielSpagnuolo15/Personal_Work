```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import os
import random
from concurrent.futures import ThreadPoolExecutor
import re
```

```python
def save_to_csv(df, filename):
    if not os.path.isfile(filename):
        df.to_csv(filename, index=False)
    else:  
        df.to_csv(filename, mode='a', header=False, index=False)
```

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
full_list = []
link_list = []
position_list = []
debut_year_list = []
final_year_list = []

def scrape_letter(letter):
    driver = None
    try:
        driver = webdriver.Chrome(options=options)
        url = f"https://www.hockey-reference.com/players/{letter}/"
        driver.get(url)
        time.sleep(5)

        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')

        nhl_players = soup.select('div#div_players p.nhl a')
        
        player_names = [player.text for player in nhl_players]
        player_link = [player['href'] for player in nhl_players]
        positions_and_years = [player.find_parent('p').get_text() for player in nhl_players]
        
        # Updated regular expression to handle multiple positions
        debut_years, final_years, positions = zip(*[
            (match.group(1), match.group(2), match.group(3)) if (match := re.search(r'\((\d{4})-(\d{4}), ([\w/]+)\)', item)) 
            else (None, None, None)
            for item in positions_and_years
        ])
        
        time.sleep(random.uniform(35, 45))
        return player_names, player_link, positions, debut_years, final_years

    except Exception as e:
        print(f"Error scraping letter {letter}: {e}")
        return [], [], [], [], []

    finally:
        if driver is not None:
            driver.quit()

with ThreadPoolExecutor(max_workers=5) as executor:
    results = executor.map(scrape_letter, letters)

for player_names, player_link, positions, debut_years, final_years in results:
    full_list.extend(player_names)
    link_list.extend(player_link)
    position_list.extend(positions)
    debut_year_list.extend(debut_years)
    final_year_list.extend(final_years)
```

```python
print(full_list[:10])
print(link_list[:10])
print(position_list[:10])
print(debut_year_list[:10])
print(final_year_list[:10])
```

```python
df = pd.DataFrame({'Player Names': full_list, 'Player Links': link_list, 'Position': position_list, 'Debut Year': debut_year_list,
                   'Final Year to Date': final_year_list})
df.head(20)
```
