```python
import pandas as pd
import time
from concurrent.futures import ThreadPoolExecutor
import os
import random
import time
import urllib.error
import socket
import datetime
import requests
from bs4 import BeautifulSoup
```

```python
def save_to_csv(df, filename):
    if not os.path.isfile(filename):
        df.to_csv(filename, index=False)
    else:  
        df.to_csv(filename, mode='a', header=False, index=False)
```

```python
start_year = 2011
current_year = datetime.datetime.now().year
player_names = []
player_salaries = []
years = []

for year in range(start_year, current_year + 1):
    url = f'https://www.spotrac.com/nhl/rankings/player/_/year/{year}/sort/cap_total'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_elements = soup.find_all('div', class_='link')
    span_elements = soup.find_all('span', class_='medium')
    text_list = [div.get_text() for div in div_elements]
    money_list = [span.get_text(strip=True).replace('\n', '').replace(',', '').replace('$', '') for span in span_elements]
    money_list = [value for value in money_list if value.isdigit()]
    num_entries = min(len(text_list), len(money_list))
    player_names.extend(text_list[:num_entries])
    player_salaries.extend(money_list[:num_entries])
    years.extend([f'{year} - {year + 1}'] * num_entries)
    time.sleep(75)

df = pd.DataFrame({'Year': years, 'Player Names': player_names, 'Player Salaries': player_salaries})
```

```python
player_money_data = r"C:\Users\gabri\Documents\player_money_data.csv"
save_to_csv(df, player_money_data)
```

```python
start_year = 2011
current_year = datetime.datetime.now().year
team_salaries = []

for year in range(start_year, current_year + 1):
    full_df = pd.read_html(f"https://www.spotrac.com/nhl/cap/_/year/{year}/sort/cap_total2")
    df2 = full_df[0]
    df2['Year'] = f'{year} - {year + 1}'
    team_salaries.append(df2)
    time.sleep(75)

teams_df = pd.concat(team_salaries, ignore_index=True)
```
