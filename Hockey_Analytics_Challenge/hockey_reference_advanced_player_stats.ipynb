```python
import pandas as pd
import time
from concurrent.futures import ThreadPoolExecutor
import os
import random
import time
import urllib.error
import socket
import datetime
import requests
from bs4 import BeautifulSoup
```

```python
def save_to_csv(df, filename):
    if not os.path.isfile(filename):
        df.to_csv(filename, index=False)
    else:  
        df.to_csv(filename, mode='a', header=False, index=False)
```

```python
names_df = pd.read_csv(r"C:\Users\gabri\Documents\nhl_players_dataset.csv")
```

```python
names_df.columns = ["Player Name", "Player Link", "Position", "Debut Year", "Final Year"]
names_df.head()
```

```python
names_df['Total Years'] = (names_df['Final Year'] - names_df['Debut Year']) + 1
names_df
```

```python
condition1 = ((names_df['Final Year'] != 2024) & (names_df['Total Years'] < 5)) | ((names_df['Final Year'] < 2008))
filtered_df = names_df[~condition1]
filtered_df
```

```python
df_goalies = filtered_df[filtered_df['Position'] == 'G']
df_goalies
```

```python
df_players = filtered_df[filtered_df['Position'] != 'G']
df_players
```

```python
player_name_list = df_players['Player Name'].tolist()
player_link_list = df_players['Player Link'].tolist()
```

```python
full_df_players_list = []
batch_size = 200

def fetch_html_with_retry(url, retries=3, delay=50):
    """Fetch HTML content with retries in case of network errors."""
    for i in range(retries):
        try:
            full_df = pd.read_html(url)
            return full_df
        except (urllib.error.URLError, socket.gaierror) as e:
            print(f"Attempt {i+1} failed: {e}")
            time.sleep(delay)
    raise Exception(f"Failed to retrieve data from {url} after {retries} attempts")

for batch_start in range(0, len(player_link_list), batch_size):
    batch = player_link_list[batch_start:batch_start + batch_size]
    for players in batch:  # Corrected to iterate over the batch
        url = f"https://www.hockey-reference.com{players}"
        try:
            full_df = fetch_html_with_retry(url)
            player_df = full_df[8]
            player_df['Player Links'] = players
            player_df['Situation'] = 'All'
            full_df_players_list.append(player_df)
            time.sleep(1)

            player_df2 = full_df[9]
            player_df2['Player Links'] = players
            player_df2['Situation'] = 'Even Strength'
            full_df_players_list.append(player_df2)
            time.sleep(1)

            player_df3 = full_df[10]
            player_df3['Player Links'] = players
            player_df3['Situation'] = 'Power Play'
            full_df_players_list.append(player_df3)
            time.sleep(1)

            player_df4 = full_df[11]
            player_df4['Player Links'] = players
            player_df4['Situation'] = 'Short-Handed'
            full_df_players_list.append(player_df4)
            time.sleep(1)

            player_df5 = full_df[12]
            player_df5['Player Links'] = players
            player_df5['Situation'] = '5-On-5'
            full_df_players_list.append(player_df5)
            time.sleep(1)

            player_df6 = full_df[13]
            player_df6['Player Links'] = players
            player_df6['Situation'] = '5-On-5 Close'
            full_df_players_list.append(player_df6)
            time.sleep(1)

            player_df7 = full_df[14]
            player_df7['Player Links'] = players
            player_df7['Situation'] = '5-On-5 Tied'
            full_df_players_list.append(player_df7)
            time.sleep(1)

        except Exception as e:
            print(f"Skipping {url} due to repeated failures.")

    time.sleep(random.randint(100, 220))  # Longer delay between batches

full_df_players = pd.concat(full_df_players_list, ignore_index=True)
```
