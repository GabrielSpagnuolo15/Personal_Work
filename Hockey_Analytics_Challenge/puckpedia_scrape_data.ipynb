```python
import pandas as pd
import time
import os
import random
import requests
```

```python
def save_to_csv(df, filename):
    if not os.path.isfile(filename):
        df.to_csv(filename, index=False)
    else:  
        df.to_csv(filename, mode='a', header=False, index=False)
```

```python
nhl_teams = [
    "Anaheim-Ducks",
    "Boston-Bruins",
    "Buffalo-Sabres",
    "Calgary-Flames",
    "Carolina-Hurricanes",
    "Chicago-Blackhawks",
    "Colorado-Avalanche",
    "Columbus-Blue-Jackets",
    "Dallas-Stars",
    "Detroit-Red-Wings",
    "Edmonton-Oilers",
    "Florida-Panthers",
    "Los-Angeles-Kings",
    "Minnesota-Wild",
    "Montreal-Canadiens",
    "Nashville-Predators",
    "New-Jersey-Devils",
    "New-York-Islanders",
    "New-York-Rangers",
    "Ottawa-Senators",
    "Philadelphia-Flyers",
    "Pittsburgh-Penguins",
    "San-Jose-Sharks",
    "Seattle-Kraken",
    "St-Louis-Blues",
    "Tampa-Bay-Lightning",
    "Toronto-Maple-Leafs",
    "utah-hc",
    "Vancouver-Canucks",
    "Vegas-Golden-Knights",
    "Washington-Capitals",
    "Winnipeg-Jets"
]

df = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()

for team in nhl_teams:
    url = f"https://puckpedia.com/team/{team}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
    response = requests.get(url, headers=headers)
    tables = pd.read_html(response.text)
    df_forwards = tables[7].iloc[:-1]
    df_forwards['Team'] = team
    df = pd.concat([df, df_forwards], ignore_index=True)
    time.sleep(random.randint(3, 5))
    
    df2_defence = tables[8].iloc[:-1]
    df2_defence['Team'] = team
    df2 = pd.concat([df2, df2_defence], ignore_index=True)
    time.sleep(random.randint(3, 5))
    
    df3_goalie = tables[9].iloc[:-1]
    df3_goalie['Team'] = team
    df3 = pd.concat([df3, df3_goalie], ignore_index=True)
    time.sleep(random.randint(3, 5))

    time.sleep(random.randint(20, 35))
```

```python
df['Player'] = df['Player'].astype(str)
extracted_data = df['Player'].str.extract(r'([^ ]+), ([^ ]+)  age(\d+)  pos([^ ]+) shoots([a-zA-Z]+)(?: (LTIR|IR))?')
extracted_data.columns = ['LastName', 'FirstName', 'Age', 'Position', 'Shoots', 'InjuryStatus']
extracted_data['InjuryStatus'] = extracted_data['InjuryStatus'].fillna('n/a')
df = pd.concat([df, extracted_data], axis=1)
df = df.drop(columns=['Player'])
df.head()
```

```python
for year in ['2024-25', '2025-26', '2026-27', '2027-28']:
    df[year + ' Cap Hit'] = df[year].str.extract(r'\$([\d,]+)').replace(',', '', regex=True).astype(float)
    df[year + ' Cap Hit'] = df[year + ' Cap Hit'].fillna(0)

df = df.drop(columns=['2024-25', '2025-26', '2026-27', '2027-28', 'Unnamed: 5'])
df.head()
```
